# Write your first App !

### First, import flask

```python
from flask import Flask
```

### Then, let's create a flask application

```python
app = Flask(__name__)
```

### Now let's create our first page

```python
def index():
    return "Hello World !"
```

### Let's route it to the "/index" uri

```python
@app.route('/index')
def index():
    return "Hello World !"
```

### Run the app !

```python
app.run(port=5000, debug=True)
```

# Whole code

```python
from flask import Flask

app = Flask(__name__)

@app.route("/index")
def index():
    return "Hello World"

if __name__ == "__main__":
    app.run(port=5000, debug=True)
```
